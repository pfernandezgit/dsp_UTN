```python
import numpy as np
import matplotlib.pyplot as plt

#definición de funciones
def  mi_funcion_sen( vmax , dc , ff , ph, nn , fs ):
    
    t_start = ph/ ( ff * 2 * np.pi ) # Seg
    t_sim = nn / fs
    t = np.arange( start = t_start,stop = t_sim + t_start, step = 1 / fs  ) #Seg
    
    x = vmax * np.sin( 2 * np.pi * ff* t) + dc #Volts
    
    return t, x

 #duty expresado en porcentual
def mi_funcion_squear(  vmax, vmin, duty, ff, nn, fs ):
    t_sim = nn / fs
    t = np.arange(0, t_sim, 1/fs)
    
    x = vmax*np.ones([nn,1]) 
    x[(duty*nn)//100:] = vmin
    
    return t,x
```

A continuación se definen los parametros de prueba de las funciones

```python
delta_f = 1 #Hz
fs = 500 #Hz
N = fs # para que delta_f sea 1 (elección de diseño)
fx = 1 #Hz
ax = 1 #V
offset = 0
fase = 0
# Parametros específicos para señal cuadrada
vmax = 3
vmin = 1
duty = 30

#################
#fs - frecuencia de muestreo [Hz]
#delta_f - resolución espectral [Hz]
#fx - frecuencia de la señal [Hz]
#ax - Amplitud de la señal senoidal [V]
#offset - tension de offset [V]
#fase - expresada en radianes
#vmax - Valor máximo de señal cuadrada [V]
#vmin - valor minimo de señal cuadrada [V]
#duty - expresado en porcentual
```

Prueba de la función mi_funcion_sen

```python
tt, xx = mi_funcion_sen( ax, offset, fx, fase, N, fs)
plt.plot(tt,xx)
plt.xlabel('Segundos [s]')
plt.ylabel('Tensión [V]')
```

```python
tt, xx = mi_funcion_squear(vmax,vmin,duty,fx,N,fs)
plt.plot(tt,xx)
plt.xlabel('Segundos [s]')
plt.ylabel('Tensión [V]')
```
